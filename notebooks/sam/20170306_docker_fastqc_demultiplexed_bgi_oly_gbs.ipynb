````python
%%bash
for html in /owl_web/nightingales/O_lurida/20160223_gbs/1[1HNS]*.html
    do
    printf "- [%s](http://owl.fish.washington.edu/nightingales/O_lurida/20160223_gbs/%s)" "$html" "$html"
    done
````

````python
%%bash
for html in /owl_web/nightingales/O_lurida/20160223_gbs/1[1HNS]*.html
    do
    printf "\- [%s](http://owl.fish.washington.edu/nightingales/O_lurida/20160223_gbs/%s)" "$html" "$html"
    done
````

That didn't work, as the backslash is interpreted literally.

I'm doing this incorrectly. I should be specifying the print format and then providing printf with the junk I want printed...

The code below tells printf to print four strings in a row, followed by a newline. The %s indicates a string should printed in this position. The list of possible strings (each item in the list is contained in quotes and separated from other items by a space) is supplied after the format layout. The list is "consumed" in order (i.e. The first %s uses the first string in the list. That string is then removed from the list and the list is shifted to the left).

````python
%%bash
for html in /owl_web/nightingales/O_lurida/20160223_gbs/1[1HNS]*.html
    do
    printf "%s%s%s%s\n" "- " "[$html]" "(http://owl.fish.washington.edu/nightingales/O_lurida/20160223_gbs/" "$html"
    done
````

Closer... But, it turns out that the for loop is storing the entire path for each file in the ```html``` variable. I only want the file name.

The command below uses parameter substitution (```${html##*/}```)to delete the path from the value stored in the ```html``` variable (i.e. it deletes everything before, and including, the last forward slash from the value stored in the ```html``` variable).

````python
%%bash
for html in /owl_web/nightingales/O_lurida/20160223_gbs/1[1HNS]*.html
    do
    printf "%s%s%s%s\n" \
    "- " "[$html]" "(http://owl.fish.washington.edu/nightingales/O_lurida/20160223_gbs/" "${html##*/}" ")"
    done
````

Much closer... But, I forgot to add an additional %s to the printf format to account for the addition of the ")" that I added to the list of strings.

````python
%%bash
for html in /owl_web/nightingales/O_lurida/20160223_gbs/1[1HNS]*.html
    do
    printf "%s%s%s%s%s\n" \
    "- " "[$html]" "(http://owl.fish.washington.edu/nightingales/O_lurida/20160223_gbs/" "${html##*/}" ")"
    done
````

OMG! So, so close. I want the output to just display the filename for the link; not the full file path. Put another parameter substitution and this should work the way I wanted...

````python
%%bash
for html in /owl_web/nightingales/O_lurida/20160223_gbs/1[1HNS]*.html
    do
    printf "%s%s%s%s%s\n" \
    "- " "[" ""${html##*/}"" "]" "(http://owl.fish.washington.edu/nightingales/O_lurida/20160223_gbs/" "${html##*/}" ")"
    done
````
